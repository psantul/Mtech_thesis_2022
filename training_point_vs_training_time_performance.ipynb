```python
import pyspark.sql.functions as func
dataset = dataset.withColumn("redshift", func.round(dataset["redshift"], 2).cast('integer'))
```

```python
from pyspark.sql.functions import col
new_data = dataset.select(*(col(c).cast("float").alias(c) for c in dataset.columns))
```

```python
cols=new_data.columns
# cols = ['modelMag_u', 'modelMag_g', 'modelMag_r']
# cols.remove("Classes")
cols.remove("redshift")
cols.remove("petroR90_zz")
cols.remove("ri")
cols.remove("iz")
cols.remove("redshift_classes")


assembler = VectorAssembler(inputCols=cols,outputCol="features")

# Now let us use the transform method to transform our dataset
assembled_data=assembler.transform(new_data)

assembled_data.select("features",'redshift_classes').show(truncate=False)
```

```python
from pyspark.ml.feature import MinMaxScaler
scaler = MinMaxScaler(inputCol="features", outputCol="scaled_Features")
# rescale each feature to range [min, max].
scaled_data = scaler.fit(assembled_data).transform(assembled_data)
```

```python
scaled_data.select("features",'redshift_classes','Scaled_features').show(truncate=False)
```

```python
scaled_data = scaled_data.select("Scaled_features","redshift_classes")
scaled_data.show()
```

```python
from pyspark.ml.classification import NaiveBayes

from pyspark.ml.classification import LinearSVC, OneVsRest
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

from pyspark.ml.classification import RandomForestClassifier

from pyspark.ml.classification import DecisionTreeClassifier
```

```python
while(i<0.99):
  iteration = iteration + 1
  train_data, test_data = scaled_data.randomSplit([i, 1-i])
  dt = RandomForestClassifier(featuresCol = 'Scaled_features', labelCol = 'redshift_classes')
  begin = time.time()
  dtModel = dt.fit(train_data)
  end = time.time()
  training_pt_rf.append((end-begin)/60)
  i = i+0.04
  print(iteration)
# predictions = dtModel.transform(test_data)
# predictions.select('Scaled_features', 'redshift', 'rawPrediction', 'probability', 'prediction').show(25)
# print(end-begin)
```
