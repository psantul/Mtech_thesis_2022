```python
from pyspark.sql.functions import col
new_data = dataset.select(*(col(c).cast("float").alias(c) for c in dataset.columns))
```

```python
new_data.printSchema()
```

```python
from pyspark.sql.functions import col, count, isnan, when
#checking for null ir nan type values in our columns
new_data.select([count(when(col(c).isNull(), c)).alias(c) for c in new_data.columns]).show()
```

```python
cols=new_data.columns
# cols = ['modelMag_u', 'modelMag_g', 'modelMag_r']
# cols.remove("Classes")
cols.remove("redshift")
assembler = VectorAssembler(inputCols=cols,outputCol="features")

# Now let us use the transform method to transform our dataset
data=assembler.transform(new_data)

data.select("features",'redshift').show(truncate=False)
```

```python
standardscaler=StandardScaler().setInputCol("features").setOutputCol("Scaled_features")
data=standardscaler.fit(data).transform(data)
```

```python
data.select("features",'redshift','Scaled_features').show(truncate=False)
```

```python
assembled_data = data.select("Scaled_features","redshift")
assembled_data.show()
```

```python
assembled_data = assembled_data.withColumnRenamed("redshift","label")
```

```python
assembled_data = assembled_data.withColumnRenamed("Scaled_features","features")
```

```python
train, test = assembled_data.randomSplit([0.7, 0.3])
```

```python
train.show()
```

```python
test.show()
```

```python
from pyspark.ml.classification import LinearSVC, OneVsRest
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
```

```python
# Define your binary classifier
lsvc = LinearSVC(maxIter=10, regParam=0.1)
```

```python
# Define the One Vs All Classifier.
ovr = OneVsRest(classifier=lsvc)
```

```python
# train the multiclass model.
ovrModel = ovr.fit(train)
```

```python
# score the model on test data.
predictions = ovrModel.transform(test)
```

```python
# Show the computed predictions and compare with the original labels
predictions.select("features", "label", "prediction").show(10)
```

```python
predictionAndLabels = predictions.select("label","prediction").rdd
```

```python
# Define the evaluator method with the corresponding metric and compute the classification error on test data
evaluator =  MulticlassClassificationEvaluator().setMetricName('accuracy')
accuracy = evaluator.evaluate(predictions) 

# Show the accuracy
print("Test accuracy = %g" % (accuracy))
```

```python
from pyspark.ml.evaluation import RegressionEvaluator
evaluator = RegressionEvaluator(labelCol="label", predictionCol="prediction", metricName="rmse")
rmse = evaluator.evaluate(predictions)
print("Root Mean Squared Error (RMSE) on test data = %g" % rmse)

evaluator = RegressionEvaluator(labelCol="label", predictionCol="prediction", metricName="mae")
mae = evaluator.evaluate(predictions)
print("Mean Absolute Eroor (MAE) on test data = %g" % mae)

```

```python
from sklearn.metrics import confusion_matrix
y_pred=predictions.select("prediction").collect()
y_orig=predictions.select("label").collect()

cm = confusion_matrix(y_orig, y_pred)
print("Confusion Matrix:")
print(cm)
```

```python
from pyspark.mllib.evaluation import MulticlassMetrics

# Instantiate metrics object
metrics = MulticlassMetrics(predictionAndLabels)

# Overall statistics
precision = metrics.precision(0.0)
recall = metrics.recall(0.0)
f1Score = metrics.fMeasure(0.0)
print("Summary Stats")
print("Precision = %s" % precision)
print("Recall = %s" % recall)
print("F1 Score = %s" % f1Score)
```
